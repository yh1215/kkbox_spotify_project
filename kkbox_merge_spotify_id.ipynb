# track

```python
import pandas as pd

k_2021 = pd.read_csv('kkbox_data/kkbox_data_2021.csv')
k_2022 = pd.read_csv('kkbox_data/kkbox_data_2022.csv')
k_2023 = pd.read_csv('kkbox_data/kkbox_data_2023.csv')
k_2024 = pd.read_csv('kkbox_data/kkbox_data_2024.csv')

df = pd.concat([k_2021, k_2022, k_2023, k_2024])
df.head()
```

```python
track = df[df['song_id'] != '#'][['artist', 'song_id']].drop_duplicates()
track.rename(columns={'song_id': 'id'}, inplace=True)

track.head()
```

```python
ttl_track = pd.read_csv('kkbox_data/tracks.csv')
ttl_track.head()
```

```python
rank_track = track.merge(ttl_track[['name', 'id']], how='inner', on='id')
rank_track.head()
```

```python
rank_track.to_csv('test_track_merge.csv', index=False)
```

# artist

```python
import pandas as pd

kkbox_artist = pd.read_csv('kkbox_data/artists.csv')
kkbox_artist
```

```python
kkbox_artist['spotify_artist'] = kkbox_artist['artist'].str.split('(').str[0].str.strip()
kkbox_artist
```

```python
# 創建一個臨時列來存儲擴展後的結果
df_expanded = kkbox_artist.assign(temp_artist=kkbox_artist['spotify_artist'].str.split(','))

# 使用 explode 函數擴展臨時列
df_expanded = df_expanded.explode('temp_artist')

# 去除空白字符
df_expanded['temp_artist'] = df_expanded['temp_artist'].str.strip()

# 用擴展後的結果替換 spotify_artist 列
df_expanded['spotify_artist'] = df_expanded['temp_artist']

# 刪除臨時列
df_expanded = df_expanded.drop('temp_artist', axis=1)
```

```python
print(df_expanded)
```

```python
df_expanded.to_csv('test_artist_merge.csv', index=False)
```
