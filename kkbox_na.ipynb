# artist

```python
mandarin_artist = mandarin['artist'].drop_duplicates()
western_artist = western['artist'].drop_duplicates()
```

```python
na_artist = pd.concat([mandarin_artist, western_artist]).to_frame(name='artist')
na_artist
```

```python
len(na_artist.drop_duplicates())
```

```python
na_artist = na_artist.drop_duplicates()

import re

def remove_last_parentheses(text):
    # 只匹配最後一個以括號開頭的部分並將其移除
    return re.sub(r'\s*\([^)]*\)$', '', text)

na_artist['spotify_artist'] = na_artist['artist'].apply(remove_last_parentheses)
```

```python
na_artist
```

```python
def unify_separators(text):
    # 將 ' & ', '、' 替換為逗號
    return re.sub(r'\s*&\s*|、', ',', text)

# 統一分隔符號
na_artist['spotify_artist'] = na_artist['spotify_artist'].apply(unify_separators)
```

```python

# 創建一個臨時列來存儲擴展後的結果
df_expanded = na_artist.assign(temp_artist=na_artist['spotify_artist'].str.split(','))

# 使用 explode 函數擴展臨時列
df_expanded = df_expanded.explode('temp_artist')

# 去除空白字符
df_expanded['temp_artist'] = df_expanded['temp_artist'].str.strip()

# 用擴展後的結果替換 spotify_artist 列
df_expanded['spotify_artist'] = df_expanded['temp_artist']

# 刪除臨時列
df_expanded = df_expanded.drop('temp_artist', axis=1)
```

```python
df_expanded
```

```python
df_expanded = pd.read_csv('na_artist.csv')
```

```python
na_artist_merge = pd.read_csv('na_artist_merge.csv')
na_artist_merge
```

```python
na_artist_merge.rename(columns={'artist': 'spotify_artist'}, inplace=True)

artist_unique = na_artist_merge.drop_duplicates(subset=['spotify_artist'])

# 然後進行合併操作
result = df_expanded.merge(na_artist_merge, how='inner', on='spotify_artist')
```
